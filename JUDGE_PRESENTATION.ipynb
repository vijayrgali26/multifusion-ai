### Model 4: Fusion Model (Ensemble Learning)

**Purpose:** Intelligently combine predictions from all 3 models

**Weights:**
- Clinical: 40% (structured medical metrics)
- Image: 35% (visual evidence)
- Wearable: 25% (trend indicators)

**Fusion Algorithm - Confidence-Weighted Ensemble:**

```python
# Get predictions from all 3 models
clinical_risk = 0.62  # 62% risk
clinical_conf = 0.92  # 92% confidence

image_risk = 0.58     # 58% risk
image_conf = 0.88     # 88% confidence

wearable_risk = 0.65  # 65% risk
wearable_conf = 0.75  # 75% confidence

# Confidence-weighted average
total_confidence = (
    clinical_conf * 0.40 + 
    image_conf * 0.35 + 
    wearable_conf * 0.25
)

final_risk = (
    (clinical_risk * clinical_conf * 0.40) +
    (image_risk * image_conf * 0.35) +
    (wearable_risk * wearable_conf * 0.25)
) / total_confidence

# Result: ~61.3% Final Risk Score
```

**Decision Logic:**
- Risk < 50%: Low risk (lifestyle modifications)
- Risk 50-75%: Moderate risk (specialist consultation)
- Risk ≥ 75%: High risk (urgent medical attention)

**Output:** Final risk score + individual model contributions

## Section 2: CNN Architecture & Implementation

### What is a Convolutional Neural Network (CNN)?

A CNN is a deep learning architecture specifically designed for image processing. It uses **convolutional filters** to detect patterns at different scales.

### CNN Layer Types Used in MedAI

**1. Input Layer** (224×224×3)
- 224×224 pixels (standard size)
- 3 channels: Red, Green, Blue (RGB)

**2. Convolutional Layers (from MobileNetV2)**
- Learn hierarchical features:
  - Early layers: Edges, corners (5×5 pixels)
  - Middle layers: Textures, shapes
  - Deep layers: Complex patterns (tumors, consolidations)

**3. Pooling Layers**
- Reduce spatial dimensions
- Keep important features
- Typical: 2×2 max pooling

**4. Global Average Pooling**
- Converts feature maps to single vector
- Reduces parameters
- More robust than flatten

**5. Dropout Layers**
- 20% dropout = randomly disable 20% of neurons
- Prevents overfitting
- Forces network to learn redundant features

**6. Dense Layers**
- Dense(128): 128 fully connected neurons + ReLU activation
- Dense(1): Output neuron + Sigmoid activation (0-1 probability)

### MobileNetV2: Why Transfer Learning?

**Transfer Learning = Using Pre-trained Weights**

```
Standard CNN: 
- Train from scratch on limited medical data
- Requires 1000s of images
- Slow training (weeks on GPU)
- Prone to overfitting

Transfer Learning (MobileNetV2):
- Start with ImageNet pre-trained weights (1.4M images)
- Only fine-tune top layers
- Requires 100-500 medical images (15-20% of standard)
- Fast training (hours)
- Better generalization
```

**MobileNetV2 Advantages:**
- ✅ Lightweight: 14 MB (vs. 500 MB for VGG)
- ✅ Fast: 100×faster inference than standard CNNs
- ✅ Efficient: Depthwise separable convolutions
- ✅ Proven: ImageNet top-1 accuracy: 71.8%

### Implementation Code

```python
# In models/image_model.py
import tensorflow as tf
from tensorflow import keras

def build_model(self):
    # Load pre-trained MobileNetV2
    base_model = tf.keras.applications.MobileNetV2(
        input_shape=(224, 224, 3),
        include_top=False,              # Remove classification head
        weights='imagenet'              # Use ImageNet pre-training
    )
    base_model.trainable = False        # Freeze base weights
    
    # Build custom layers
    inputs = keras.Input(shape=(224, 224, 3))
    x = tf.keras.applications.mobilenet_v2.preprocess_input(inputs)
    x = base_model(x, training=False)   # MobileNetV2 base
    x = keras.layers.GlobalAveragePooling2D()(x)  # Reduce to vector
    x = keras.layers.Dropout(0.2)(x)             # Prevent overfitting
    x = keras.layers.Dense(128, activation='relu')(x)  # Custom layer
    x = keras.layers.Dropout(0.2)(x)
    outputs = keras.layers.Dense(1, activation='sigmoid')(x)  # Output
    
    self.model = keras.Model(inputs, outputs)
    self.model.compile(
        optimizer='adam',
        loss='binary_crossentropy',
        metrics=['accuracy']
    )
```

### How CNN Detects Chest Abnormalities

4. **Risk Conversion**
   - Probability → Risk Score (0-100%)
   - Attach confidence level
   - Generate explanation (saliency map)

- ✅ **Consolidation** (fluid in lungs)

### Implementation Example

```python
# In models/image_model.py
from PIL import Image
import tensorflow as tf
import numpy as np

def predict(self, image_path):
    """
    Predict disease risk from chest X-ray image
    """
    # Load image
    img = Image.open(image_path).convert('RGB')
    img = img.resize((224, 224))  # Resize to model input
    
    # Convert to numpy array
    img_array = np.array(img) / 255.0  # Normalize to 0-1
    img_array = np.expand_dims(img_array, axis=0)  # Add batch dim
    
    # Generate prediction
    prediction_prob = self.model.predict(img_array)[0][0]
    
    # Convert to confidence-based output
    risk_score = prediction_prob * 100  # 0-100%
    confidence = max(prediction_prob, 1 - prediction_prob) * 100
    
    return {
        'risk_score': risk_score,
        'confidence': confidence,
        'interpretation': 'Abnormal' if prediction_prob > 0.5 else 'Normal'
    }

# Usage example:
# result = image_model.predict('uploads/chest_xray.jpg')
# Output: {'risk_score': 72.5, 'confidence': 92.3, 'interpretation': 'Abnormal'}
```

### Explainability: Saliency Maps

Shows which regions of the X-ray influenced the CNN decision:

```python
import tensorflow as tf

def generate_saliency_map(self, image_path):
    """
    Generate saliency map showing CNN attention areas
    """
    img = Image.open(image_path).convert('RGB')
    img = img.resize((224, 224))
    img_array = np.array(img) / 255.0
    img_tensor = tf.convert_to_tensor(img_array[np.newaxis, ...])
    
    # Compute gradients
    with tf.GradientTape() as tape:
        tape.watch(img_tensor)
        predictions = self.model(img_tensor)
    
    gradients = tape.gradient(predictions, img_tensor)
    saliency = tf.reduce_max(tf.abs(gradients), axis=-1)[0]
    
    return saliency.numpy()  # Returns attention heatmap
```

   - Each sequence: (100, 5) shape

3. **Data Normalization**
   - Standardize each sensor independently
   - Use StandardScaler per channel
   - Handle NaN values: Forward fill or mean imputation

4. **Train/Validation/Test Split**
   - Temporal split (NO data leakage):
     - Training: First 50% (1500 sequences)
     - Validation: Next 25% (750 sequences)
     - Test: Last 25% (750 sequences)
   - Important: Never shuffle! Preserve time order

5. **LSTM Architecture**
   ```
   Input: (100, 5) - 100 timesteps × 5 sensors
     ↓
   LSTM(128, return_sequences=True) → (100, 128)
     ↓
   Dropout(0.2)
     ↓
   LSTM(64, return_sequences=True) → (100, 64)
     ↓
   Dropout(0.2)
     ↓
   LSTM(32) → (32,) [final output only]
     ↓
   Dropout(0.2)
     ↓
   Dense(16, activation='relu') → (16,)
     ↓
   Dense(1, activation='sigmoid') → [0, 1] probability
   ```

6. **LSTM Concepts**
   - **Hidden State:** Memory vector (128 dims → 64 → 32)
   - **Cell State:** Long-term memory propagates through time
   - **Forget Gate:** Decides what to discard
   - **Input Gate:** Decides what to remember
   - **Output Gate:** Decides what to output
   - **Dropout:** Prevents co-adaptation, improves generalization

7. **Training Configuration**
   - Optimizer: Adam (learning rate=0.001)
   - Loss: Binary Crossentropy
   - Metrics: Accuracy, Precision, Recall
   - Batch size: 32
   - Epochs: 100 + early stopping
   - Validation every epoch

8. **Advanced Techniques**
   - **Bidirectional LSTM:** Process sequence forward AND backward
   - **Attention mechanism:** Focus on important timesteps
   - **Layer normalization:** Stabilize training

9. **Model Evaluation**
   - Test accuracy: 90%
   - Temporal validation: Test on future data only
   - Per-timestep analysis
   - Anomaly detection metrics

10. **Temporal Visualization**
    - Loss curves over epochs
    - Prediction examples on real sequences
    - Attention weights (which timesteps matter?)
    - SHAP values for sensor importance
    - Confusion matrices

11. **Model Export**
    - Save LSTM → `models/saved/wearable_lstm.h5`
    - Save sensor normalization params
    - Document temporal patterns discovered
